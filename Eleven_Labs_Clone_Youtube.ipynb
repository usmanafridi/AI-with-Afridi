**NOW USING THIS VOICE ID TO GENERATE TEXT-TO-SPEECH**

```python
from elevenlabs import ElevenLabs

audio_generate= client.text_to_speech.convert(
    voice_id="18UI5bom80Zj3bqojZ6M",
    output_format="mp3_44100_128",
    text="This is my first Youtube video, and I think this is going to be the beginning of something awesome!!.",
    model_id="eleven_multilingual_v2",
)
```

```python
# Save the audio file
output_path = "usman_audio_sample.mp3"
with open(output_path, "wb") as f:
    for chunk in audio_generate:
        f.write(chunk)

print(f"Audio file saved as {output_path}")
```
