**GENERATE SIMPLE TEXT-TO-SPEECH**

```python
client = ElevenLabs(api_key=API_Key)

audio_generate= client.text_to_speech.convert(
    voice_id="JBFqnCBsd6RMkjVDRZzb",
    output_format="mp3_44100_128",
    text="Hello there, how is it going John? I want to meet you next Thursday. Let me know when you are available",
    model_id="eleven_multilingual_v2",
)
```

```python
# Save the audio file
output_path = "audio_output1.mp3"
with open(output_path, "wb") as f:
    for chunk in audio_generate:
        f.write(chunk)

print(f"Audio file saved as {output_path}")
```

# ElevenLabs Model Parameters Overview

## Model ID
The most advanced model currently available is **[Multilingual v2](https://elevenlabs.io/docs/models#multilingual-v2)**.

## Language Code
Specify the desired `language_code` according to the supported languages in the ElevenLabs API.

## Voice Settings

### Stability
- Controls the consistency of the generated voice.
- **Lower values** introduce more variation and emotional range but may result in erratic speech.
- **Higher values** ensure stability but may lead to a monotonous tone.
- Recommended approach:
  - For **lively and dramatic** performances: Use a **lower** stability setting and generate multiple samples.
  - For **serious and consistent** outputs: Use a **higher** stability setting for reliable results.

### Similarity
- Dictates how closely the AI adheres to the original voice.
- **High similarity values** may retain background artifacts from poor-quality audio.
- Adjust this setting based on the clarity and quality of the input voice.

### Style
- Enhances the stylistic characteristics of the original speaker.
- **Recommended Setting:** Keep this at **0** to maintain model stability.
- Using higher values can make the voice more expressive but may lead to instability and increased latency.

### Speaker Boost
- Increases similarity to the original speaker.
- This requires higher computational power, leading to minor latency increases.
- The effects are generally **subtle** but useful for fine-tuning authenticity.

## Pronunciation Dictionary Locators
Provide any necessary pronunciation guides if specific terms need adjustments.

## Seed
- Set a **fixed seed** for reproducibility if needed.

## Previous & Next Text
- Ensure smooth transitions in dialogue by considering `previous_text` and `next_text` in sequential generations.




```python
audio_generate= client.text_to_speech.convert(


    voice_id="JBFqnCBsd6RMkjVDRZzb",
    output_format="mp3_44100_128",

    text="Hurrah!! I am so happy. I want to meet you!!!",

    model_id="eleven_multilingual_v2",
    voice_settings={
                    "stability": 0.9,
                    "similarity_boost": 0.65,
                    "use_speaker_boost": True
                },
)


```

```python
# Save the audio file
output_path = "audio_output_2.mp3"
with open(output_path, "wb") as f:
    for chunk in audio_generate:
        f.write(chunk)

print(f"Audio file saved as {output_path}")
```

```python
Audio(f"{output_path}")
```

**DOING CONTEXT AWARE AND CONSISTENt TTS**

```python
audio_generate= client.text_to_speech.convert(


    voice_id="JBFqnCBsd6RMkjVDRZzb",
    output_format="mp3_44100_128",
    text="I am so happy",

    model_id="eleven_multilingual_v2",
    voice_settings={
                    "stability": 0.9,
                    "similarity_boost": 0.65,
                    "use_speaker_boost": True
                },
    previous_text= "Hurrah!!",
    next_text= "I want to meet you!!!",
    seed=100
)
```

```python
# Save the audio file
output_path = "audio_output_4.mp3"
with open(output_path, "wb") as f:
    for chunk in audio_generate:
        f.write(chunk)

print(f"Audio file saved as {output_path}")
```
